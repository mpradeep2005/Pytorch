```python
import torch
```

Tensor datatype is one of the 3 big issuse we run into with Pytorch & deep learing

1.Tensor not right datatype

2.tensor not right shape

3.Tensor not on the right device

```python
float_32=torch.tensor([3,3,2],#these three are important
                      device="cpu",
                      dtype=torch.float32,#full precision point(stable))
                      requires_grad=False)
float_32.dtype
```

```python
float_16=torch.tensor(data=[3,4,2],
                      dtype=torch.float16 #half tensor (speed )
                      )
float_16
```

typecasting

```python
float_16=float_32.type(torch.float16)
float_16.dtype
```

```python
temp=float_16 * float_32
temp.dtype
```

# Getting info about a datatype
1.Tensor not right datatype - tensor.dtype

2.tensor not right shape - tensor.shape ,tensor.size()

3.Tensor not on the right device - tensor.device

```python
rand=torch.rand(2,3,4)
print(rand)
print(f"size:{rand.size()}")
print(f"datatype:{rand.dtype}")
print(f"device:{rand.device}")
```
