# Matric multiplication
1. scalar multiplication(Element wise)
2. dot product

### There are two things to satisfy in matrix multiplication
1. Matrix multiplication requires the number of columns in the first matrix to equal the number of rows in the second.
>  The inner dimension should match `(3,2) @ (2,3)`
2. The result of the matrix multiplication is the dimension of the outer matrix
> `(3,10) @ (10,3) = (3,3)`

```python
%%time
tensor1=torch.tensor([[1,2,3]])
tensor2=torch.tensor([1,2,3])
tensor=torch.matmul(tensor1,tensor2) # we can also use @ instead of torch.matmul()
tensor
```

## If the dimension is not matching we can use Transpose ( T )

```python
torch.mm(torch.rand(3,2),torch.rand(3,2).T) # we can use mm instead os matmul
```
